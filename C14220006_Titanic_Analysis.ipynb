```python
# 1. Import Library
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# 2. Load Dataset Titanic
url = "https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv"
df = pd.read_csv(url)

print("Ukuran dataset:", df.shape)
print("\nInfo dataset:")
print(df.info())
print("\nJumlah missing values:")
print(df.isnull().sum())
```

```python
# --- 3. Analisis Deskriptif ---
print("\nStatistik deskriptif variabel numerik:")
print(df.describe().T)
```

```python
# Distribusi Survived
sns.countplot(x="Survived", data=df, palette="Set2")
plt.title("Distribusi Survival (0 = Meninggal, 1 = Selamat)")
plt.show()

# Distribusi Pclass
sns.countplot(x="Pclass", data=df, palette="Set3")
plt.title("Distribusi Kelas Tiket (Pclass)")
plt.show()

# Distribusi Usia
plt.figure(figsize=(8,4))
sns.histplot(df["Age"].dropna(), bins=30, kde=True)
plt.title("Distribusi Usia Penumpang")
plt.show()

# Distribusi Fare
plt.figure(figsize=(8,4))
sns.histplot(df["Fare"], bins=30, kde=True)
plt.title("Distribusi Harga Tiket")
plt.show()
```

```python
# 4. Analisis Survival
# Gender vs Survival
sns.countplot(x="Sex", hue="Survived", data=df, palette="Set1")
plt.title("Survival berdasarkan Gender")
plt.show()
```

```python
# Pclass vs Survival
sns.countplot(x="Pclass", hue="Survived", data=df, palette="Set3")
plt.title("Survival berdasarkan Pclass")
plt.show()
```

```python
# Usia vs Survival
plt.figure(figsize=(8,4))
sns.kdeplot(data=df[df["Survived"]==0]["Age"].dropna(), label="Meninggal", shade=True)
sns.kdeplot(data=df[df["Survived"]==1]["Age"].dropna(), label="Selamat", shade=True)
plt.title("Distribusi Usia: Selamat vs Meninggal")
plt.legend()
plt.show()
```

```python
# Fare vs Survival
plt.figure(figsize=(8,4))
sns.boxplot(x="Survived", y="Fare", data=df, palette="Set2")
plt.title("Distribusi Harga Tiket: Selamat vs Meninggal")
plt.show()
```

```python
# Embarked vs Survival
sns.countplot(x="Embarked", hue="Survived", data=df, palette="Set2")
plt.title("Survival berdasarkan Pelabuhan (Embarked)")
plt.show()
```

```python
# Korelasi antar variabel numerik
plt.figure(figsize=(10,6))

# hanya ambil kolom numerik
numeric_df = df.select_dtypes(include=['int64', 'float64'])

sns.heatmap(numeric_df.corr(), annot=True, cmap="Reds", fmt=".2f")
plt.title("Heatmap Korelasi Variabel Numerik")
plt.show()
```

```python
# 6. Survival Rate (angka persentase)
print("\nSurvival Rate berdasarkan Gender:")
print(df.groupby("Sex")["Survived"].mean())

print("\nSurvival Rate berdasarkan Pclass:")
print(df.groupby("Pclass")["Survived"].mean())

print("\nSurvival Rate berdasarkan Embarked:")
print(df.groupby("Embarked")["Survived"].mean())

print("\nRata-rata usia penumpang selamat vs tidak selamat:")
print(df.groupby("Survived")["Age"].mean())
```
