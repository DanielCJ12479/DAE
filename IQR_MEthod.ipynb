```python
# Perbaikan dan versi lebih tahan banting untuk baca CSV + cleaning & plotting
import os, io, csv, re
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from google.colab import files

# --- 0. Nama file / Upload jika belum ada ---
file_path = "/content/TGM 2020-2023_eng.csv"
if not os.path.exists(file_path):
    print("File tidak ditemukan di /content â silakan upload file CSV sekarang.")
    uploaded = files.upload()  # akan munculkan dialog upload
    if not uploaded:
        raise SystemExit("Tidak ada file diupload.")
    # ambil file pertama yang diupload
    file_path = list(uploaded.keys())[0]
print(f"Memakai file: {file_path}\n")

# --- 1. Tampilkan sample (bantu diagnosis) ---
with open(file_path, "rb") as f:
    raw = f.read(5000)  # lihat beberapa KB pertama
try:
    sample_text = raw.decode("utf-8")
    used_encoding = "utf-8"
except Exception:
    sample_text = raw.decode("latin1", errors="replace")
    used_encoding = "latin1"
print(f"Deteksi encoding (sample): {used_encoding}")
print("---- 2000 karakter pertama dari file (untuk diagnosis) ----")
print(sample_text[:2000])
print("-----------------------------------------------------------\n")

# --- 2. Coba deteksi delimiter dengan csv.Sniffer ---
detected_sep = None
try:
    sniffer = csv.Sniffer()
    dialect = sniffer.sniff(sample_text)
    detected_sep = dialect.delimiter
    print(f"Delimiter terdeteksi oleh csv.Sniffer: {repr(detected_sep)}")
except Exception:
    print("Gagal otomatis mendeteksi delimiter (csv.Sniffer). Akan mencoba beberapa kemungkinan.\n")

# --- 3. Fungsi pembaca yang tahan banting (beberapa fallback) ---
def safe_read_csv(path, **kwargs):
    """Panggil pd.read_csv, fallback kalau pandas mengeluh tentang kwarg (mis. on_bad_lines)."""
    try:
        return pd.read_csv(path, **kwargs)
    except TypeError as e:
        # bisa jadi pandas versi lama tidak mendukung on_bad_lines
        kwargs2 = {k:v for k,v in kwargs.items() if k != "on_bad_lines"}
        return pd.read_csv(path, **kwargs2)

# daftar delimiter + header options untuk dicoba
delim_candidates = [detected_sep, ",", ";", "\t", "|", None]
header_candidates = [0, None]  # header=0 = baris pertama header, None = tidak ada header

read_attempts = []
df = None
for header in header_candidates:
    for sep in delim_candidates:
        if sep is None:
            kwargs = {"header": header, "encoding": used_encoding, "skipinitialspace": True, "engine": "python", "on_bad_lines": "warn"}
        else:
            kwargs = {"sep": sep, "header": header, "encoding": used_encoding, "skipinitialspace": True, "engine": "python", "on_bad_lines": "warn"}
        try:
            df_try = safe_read_csv(file_path, **kwargs)
            # jika berhasil dan punya >0 kolom, berhenti
            if df_try is not None and df_try.shape[1] > 0:
                df = df_try
                print(f"\nBerhasil membaca: sep={repr(sep)} header={header} -> shape={df.shape}")
                break
        except pd.errors.ParserError as e:
            read_attempts.append((sep, header, "ParserError", str(e)))
        except Exception as e:
            read_attempts.append((sep, header, type(e).__name__, str(e)))
    if df is not None:
        break

# fallback: coba encoding latin1 tanpa on_bad_lines
if df is None:
    try:
        df = pd.read_csv(file_path, encoding="latin1")
        print(f"\nFallback: berhasil membaca dengan encoding latin1 (default sep). shape={df.shape}")
    except Exception as e:
        print("\nGagal membaca file dengan semua percobaan. Ringkasan error yang tercatat:")
        for r in read_attempts[:10]:
            print(r)
        raise SystemExit("Stop: silakan periksa file CSV (delimiter/quoting/encoding) atau kirim 10 baris pertama agar saya bantu analisis lebih lanjut.")

# --- 4. Jika file terbaca, tampilkan 10 baris pertama dan kolom ---
print("\nPreview dataframe (head):")
display(df.head(10))
print("\nKolom yang ada:", df.columns.tolist())

# --- 5. Proses numerik (sesuaikan dengan kode awal) ---
numeric_cols_candidate = [
    "Reading Frequency per week",
    "Number of Readings per Quarter",
    "Daily Reading Duration (in minutes)",
    "Internet Access Frequency per Week",
    "Daily Internet Duration (in minutes)",
    "Tingkat Kegemaran Membaca (Reading Interest)",
]
numeric_cols = [col for col in numeric_cols_candidate if col in df.columns]

if not numeric_cols:
    print("Tidak ditemukan kolom numerik yang diharapkan dalam file. Kolom yang ada:")
    print(df.columns.tolist())
    # hentikan atau lanjut sesuai kebutuhan
else:
    # ubah koma desimal -> titik, convert ke numeric
    for col in numeric_cols:
        df[col] = pd.to_numeric(df[col].astype(str).str.replace(",", ".", regex=False).str.strip(), errors="coerce")

    print("\nStatistik deskriptif awal:")
    print(df[numeric_cols].describe().T)

    # --- 6. IQR outlier removal & plotting ---
    df_clean = df.copy()
    print("\n--- Memulai Proses Penghapusan Outlier (IQR -> ganti dengan NaN) ---")
    for col in numeric_cols:
        non_na = df_clean[col].dropna()
        if non_na.empty:
            print(f"'{col}' -> Kolom kosong, dilewati.")
            continue

        q1, q3 = non_na.quantile(0.25), non_na.quantile(0.75)
        iqr = q3 - q1
        lower_bound = q1 - 1.5 * iqr
        upper_bound = q3 + 1.5 * iqr
        outlier_count = ((non_na < lower_bound) | (non_na > upper_bound)).sum()
        print(f"'{col}' -> Batas: <{lower_bound:.2f} atau >{upper_bound:.2f}. Outlier ditemukan: {int(outlier_count)}")

        safe_col_name = re.sub(r'\W+', '_', col).strip('_').lower()

        plt.figure(figsize=(10, 4))
        # Plot SEBELUM
        plt.subplot(1, 2, 1)
        df[col].dropna().hist(bins=30)
        plt.title(f'Sebelum: {col}', fontsize=10)
        plt.xlabel(''); plt.ylabel('Count')

        # Ganti outlier dengan NaN (gunakan loc agar tidak SettingWithCopyWarning)
        mask = ~df_clean[col].between(lower_bound, upper_bound)
        df_clean.loc[mask, col] = np.nan

        # Plot SESUDAH
        plt.subplot(1, 2, 2)
        df_clean[col].dropna().hist(bins=30)
        plt.title(f'Sesudah: {col}', fontsize=10)
        plt.xlabel(''); plt.ylabel('Count')

        plt.tight_layout()
        fn = f"hist_comparison_{safe_col_name}.png"
        plt.savefig(fn)
        plt.show()
        plt.close()
        print(f"Saved: {fn}\n")

    # --- 7. Simpan hasil akhir ---
    df_final = df_clean.copy()
    print("\n--- Hasil Akhir ---")
    print(f"Jumlah baris awal: {len(df)}")
    print(f"Jumlah baris setelah proses: {len(df_final)} (baris tidak dihapus, hanya outlier diganti NaN)")
    print("\nStatistik deskriptif setelah outlier dibersihkan:")
    print(df_final[numeric_cols].describe().T)

    output_csv = "TGM_2020-2023_clean_full.csv"
    df_final.to_csv(output_csv, index=False)
    try:
        files.download(output_csv)
        print(f"\nClean Data disimpan dan siap diunduh: '{output_csv}'")
    except Exception:
        print(f"\nFile tersimpan sebagai '{output_csv}' di runtime (unduhan otomatis gagal).")
```
